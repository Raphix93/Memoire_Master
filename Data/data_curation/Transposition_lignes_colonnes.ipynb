```python
import pandas as pd
import openpyxl
```

```python
# 1. Chargement du fichier Excel
fichier_source = "02_DEF_Schweizerische_Archivstatistik_2013-2024_publiziert.xlsx"
df = pd.read_excel(fichier_source, engine='openpyxl')
```

```python
# 2. Supprimer les colonnes entièrement vides
df = df.dropna(axis=1, how='all')
```

```python
# 2. Affichage du DataFrame original (optionnel)
print("DataFrame original :")
display(df)
```

```python
# 3. Transposition des données (lignes deviennent colonnes)
df_transposé = df.transpose()
```

```python
# 4. Utiliser la première ligne transposée comme noms de colonnes
df_transposé.columns = df_transposé.iloc[0]
df_transposé = df_transposé[1:]
```

```python
# 5. Affichage du DataFrame transposé (optionnel)
print("DataFrame transposé :")
display(df_transposé)
```

```python
# 5. Récupérer l’index (archives) et le transformer en vraie colonne
df_transposé = df_transposé.reset_index().rename(columns={'index': 'archives'})
```

```python
print(df_transposé.columns)
```

```python
# 6. Nettoyage de la colonne 'archives' : suppression de tout ce qu’il y a avant (et y compris) le ":"
df_transposé['archives'] = (
    df_transposé['archives']
    .astype(str)
    .str.replace(r'^.*?:', '', regex=True)  # supprime tout jusqu’au premier ":"
    .str.replace(r"[0-9\s.]", "", regex=True)  # supprime chiffres, espaces, points
    .str.strip()
)

```

```python
# 7. Affichage de vérification
print("✅ Colonnes disponibles :", df_transposé.columns.tolist())
print("🔍 Exemple de valeurs dans 'archives' :")
display(df_transposé[['archives']].head(10))
```

```python
# 6. Sauvegarde dans un nouveau fichier Excel
fichier_sortie = "statistiques_archives_suisses_transpose_2.xlsx"
df_transposé.to_excel(fichier_sortie, index=True, engine='openpyxl')

print(f"Fichier transposé enregistré sous : {fichier_sortie}")
```
