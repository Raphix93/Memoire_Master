# Visualisation en radar des compétences clés de l'archiviste

Ce notebook contient le code utilisé pour générer la visualisation sous forme de radar de la place des compétences numériques au sein de compétences générales clés pour l'archiviste. Toutes les versions du code sont conservées dans ce notebook. La visualisation est disponible contextualisé dans le rapport de l'étude.

```python
import matplotlib.pyplot as plt
import numpy as np

# Données
labels = [
    "Compétences rédactionnelles et analytiques",
    "Aptitudes relationnelles et travail en équipe",
    "Compétences numériques générales",
    "Gestion et conduite de projets",
    "Connaissance du cadre légal et réglementaire",
    "Éthique et déontologie professionnelle"
]

values = [89.2, 87, 75, 74, 71, 77]

# Boucler les valeurs pour fermer le graphique radar
values += values[:1]  # seulement les valeurs
angles = np.linspace(0, 2 * np.pi, len(values), endpoint=False).tolist()

# Création du radar
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.3)

# Étiquettes (pas de duplication ici)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10, ha='center')

# Échelle
ax.set_rlabel_position(0)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(["20", "40", "60", "80", "100"], color="grey", size=8)
ax.set_ylim(0, 100)

plt.title("Compétences clés des archivistes (moyennes en %)", size=14, pad=20)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Données (6 branches)
labels = [
    "Compétences rédactionnelles et analytiques",
    "Aptitudes relationnelles et travail en équipe",
    "Compétences numériques générales",
    "Gestion et conduite de projets",
    "Connaissance du cadre légal et réglementaire",
    "Éthique et déontologie professionnelle"
]

values = [87, 88, 76, 72.5, 72.5, 80]
values += values[:1]  # pour fermer le graphique

# Calcul des angles pour chaque branche
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]  # pour fermer le graphique

# Création du radar
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))

# Orientation : démarrer en haut
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Tracer les données
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.3)

# Affichage des labels
ax.set_xticks(angles[:-1]) 
ax.set_xticklabels(labels, fontsize=6, fontweight='bold')

# Échelle en pourcentage
ax.set_rlabel_position(0)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(["20", "40", "60", "80", "100"], color="grey", size=8)
ax.set_ylim(0, 100)

# Affichage des valeurs sur le graphique
for angle, value in zip(angles, values):
    ax.text(angle, value + 2, f"{value}", ha='center', va='center', fontsize=7, fontweight='bold', color='black')

# Légende à droite
legend_text = (
    "Légende des scores :\n"
    "80–100  →  Essentielle\n"
    "60–80   →  Importante\n"
    "40–60   →  Utile\n"
    "20–40   →  Marginale\n"
    "0–20    →  Pas nécessaire"
)
plt.text(1.15, 0.5, legend_text, transform=ax.transAxes,
         fontsize=9, va='center', family='monospace')

# Source
plt.figtext(0.01, 0.01, "Source : Enquête sur l'évolution numérique en archives, 2025", 
            ha="left", fontsize=6)

plt.tight_layout()
plt.show()
```
