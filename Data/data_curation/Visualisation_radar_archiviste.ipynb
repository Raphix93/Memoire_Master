```python
import matplotlib.pyplot as plt
import numpy as np

# Donnes (6 branches)
labels = [
    "Comptences rdactionnelles et analytiques",
    "Aptitudes relationnelles et travail en quipe",
    "Comptences numriques gnrales",
    "Gestion et conduite de projets",
    "Connaissance du cadre lgal et rglementaire",
    "thique et dontologie professionnelle"
]

values = [89.2, 87, 75, 74, 71, 77]
values += values[:1]  # pour fermer le graphique

# Calcul des angles pour chaque branche
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]  # pour fermer le graphique

# Cration du radar
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))  # largir pour la lgende

# Orientation : dmarrer en haut
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Tracer les donnes
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.3)

# Affichage des labels
ax.set_xticks(angles[:-1]) 
ax.set_xticklabels(labels, fontsize=6, fontweight='bold')

# chelle en pourcentage
ax.set_rlabel_position(0)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(["20", "40", "60", "80", "100"], color="grey", size=8)
ax.set_ylim(0, 100)

# Lgende  droite du graphique
legend_text = (
    "Lgende des scores :\n"
    "80100    Essentielle\n"
    "6080     Importante\n"
    "4060     Utile\n"
    "2040     Marginale\n"
    "020      Pas ncessaire"
)
# Position dans lespace de la figure (transAxes = 01)
plt.text(1.15, 0.5, legend_text, transform=ax.transAxes,
         fontsize=9, va='center', family='monospace')

plt.figtext(0.01, 0.01, "Source : Enqute sur l'volution numrique en archives, 2025", ha="left", fontsize=6)

plt.tight_layout()
plt.show()
```
