```python
import matplotlib.pyplot as plt
import numpy as np

# Données (6 branches)
labels = [
    "Compétences rédactionnelles et analytiques",
    "Aptitudes relationnelles et travail en équipe",
    "Compétences numériques générales",
    "Gestion et conduite de projets",
    "Connaissance du cadre légal et réglementaire",
    "Éthique et déontologie professionnelle"
]

values = [89.2, 87, 75, 74, 71, 77]
values += values[:1]  # pour fermer le graphique

# Calcul des angles pour chaque branche
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]  # pour fermer le graphique

# Création du radar
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))  # élargir pour la légende

# Orientation : démarrer en haut
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Tracer les données
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.3)

# Affichage des labels
ax.set_xticks(angles[:-1]) 
ax.set_xticklabels(labels, fontsize=6, fontweight='bold')

# Échelle en pourcentage
ax.set_rlabel_position(0)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(["20", "40", "60", "80", "100"], color="grey", size=8)
ax.set_ylim(0, 100)

# Légende à droite du graphique
legend_text = (
    "Légende des scores :\n"
    "80–100  →  Essentielle\n"
    "60–80   →  Importante\n"
    "40–60   →  Utile\n"
    "20–40   →  Marginale\n"
    "0–20    →  Pas nécessaire"
)
# Position dans l’espace de la figure (transAxes = 0–1)
plt.text(1.15, 0.5, legend_text, transform=ax.transAxes,
         fontsize=9, va='center', family='monospace')

plt.figtext(0.01, 0.01, "Source : Enquête sur l'évolution numérique en archives, 2025", ha="left", fontsize=6)

plt.tight_layout()
plt.show()
```
